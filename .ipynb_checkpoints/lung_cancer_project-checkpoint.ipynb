## PROBLEM STATEMENT / OBJECTIVES OF LUNG CANCER CASE STUDY

Lung cancer is one of the leading causes of cancer-related deaths globally, with multiple risk factors contributing to its development. This project aims to analyze a lung cancer dataset to identify key demographic and behavioral factors associated with the disease. The analysis will focus on answering the following questions:

1. Demographics:
   - What is the age distribution of the patients?
   - How does lung cancer prevalence vary across different age groups (Young, Middle-aged, Senior)?

2. Risk Factors:
   - Are there any noticeable trends between smoking habits and lung cancer presence?
   - Is there a correlation between alcohol consumption and lung cancer?

3. Health Indicators:
   - How do other symptoms relate to lung cancer cases?

4. Comparative Analysis:
   - Do older patients exhibit different symptoms compared to younger ones?
   - Are certain symptoms more common in patients diagnosed with lung cancer?

5. Overall Insights:
   - What are the key indicators or features that stand out between those diagnosed with lung cancer and those who arenâ€™t?

The insights gained from this analysis can contribute to understanding the factors associated with lung cancer, potentially aiding early detection and intervention strategies.

```python
import pandas as pd
import matplotlib.pyplot
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#load the dataset
df = pd.read_csv('Lung Cancer Dataset.csv')
```

```python
#metadata
df.info()
```

```python
df.head() #first 5 rows of the dataset
```

```python
## checking for duplicates
df.duplicated().sum()
```

```python
# summary statistics
df.describe()
```

```python
## checking missing values
df.isnull().sum()
```

```python
df.isnull().describe()
#no null values in our dataset
```

```python
#Data Transformation using lambda functions
df['Age_group'] = df['AGE'].apply(lambda x: 'Young' if x < 40 else 'Middle-aged' if x < 60 else 'Senior')
```

```python
df.head()
```

```python
#Group by analysis
print(df.groupby('Age_group').size())
```

```python
# summary statistics of group by 
print(df.groupby('Age_group').mean(numeric_only = True))
#calculates the mean for each numerical column in each age group
```

```python
# Convert 'LUNG_CANCER' to numeric for aggregation
df['PULMONARY_DISEASE'] = df['PULMONARY_DISEASE'].apply(lambda x: 1 if x == 'YES' else 0)
```

```python
df.head()
```

## 1. Demographics:

```python
#Demographics Analysis - Age Distribution
plt.hist(df['AGE'], edgecolor = 'black', bins=20, color='blue', alpha=0.5 )
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
plt.title('AGE DISTRIBUTION')
plt.grid(True)
plt.show()
```

The age distribution of the patients shows that the ages are spread across a range, with noticeable peaks in some age groups. From the histogram:

- The age range is roughly between 30 to 85 years.

- The distribution appears fairly uniform, with no extreme skewness.

- There are slightly higher frequencies around ages 50, 60, and 70, indicating that more patients fall into those age brackets.

```python
#Demographics Analysis - Lung Cancer Prevalence by Age Group
age_group_cancer = df.groupby('Age_group')['PULMONARY_DISEASE'].mean().reset_index()
sns.barplot(x='Age_group', y='PULMONARY_DISEASE', data=age_group_cancer, palette='Blues_r', edgecolor='black')
plt.title('Lung Cancer Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Lung Cancer Rate')
plt.grid(axis='y')
plt.show()
```

- The age distribution is relatively uniform across different age groups, with patients ranging mostly between 30 and 80 years old.

- The prevalence of lung cancer is highest among middle-aged patients, slightly lower for seniors, and lowest for young patients.

## 2. Risk Factors:

```python
# Multi-factor Analysis - Age, Smoking, Alcohol vs Lung Cancer
plt.figure(figsize=(12, 8))
sns.set_palette('Set2')
sns.lineplot(data=df, x='Age_group', y='PULMONARY_DISEASE', hue='ALCOHOL_CONSUMPTION', style='SMOKING', markers=True, dashes=False)
plt.title('Lung Cancer Rate by Age, Smoking, and Alcohol Consumption')
plt.xlabel('Age Group')
plt.ylabel('Lung Cancer Rate')
plt.legend(title='Alcohol Consumption', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=1)
plt.show()
```

```python
# Multi-factor Analysis - Age, Gender, Smoking vs Lung Cancer
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Age_group', y='PULMONARY_DISEASE', hue='SMOKING', style='GENDER', markers=True, dashes=False)
plt.title('Lung Cancer Rate by Age, Gender, and Smoking Status')
plt.xlabel('Age Group')
plt.ylabel('Lung Cancer Rate')
plt.legend(title='Smoking Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
## smoking vs lung cancer 
Smoking_cancer = df.groupby('SMOKING')['PULMONARY_DISEASE'].value_counts(normalize=True).unstack().fillna(0)
Smoking_cancer.plot(kind = 'bar', stacked=False, figsize=(12,6), cmap='coolwarm' )
plt.xlabel('Smoking Status')
plt.ylabel('Proportion')
plt.title('LUNG CANCER PROPORTION BY SMOKING STATUS')
plt.show()
```

```python
## Alcohol Consumption vs lung cancer 
Alcohol_cancer = df.groupby('ALCOHOL_CONSUMPTION')['PULMONARY_DISEASE'].value_counts(normalize=True).unstack().fillna(0)
Alcohol_cancer.plot(kind = 'bar', stacked=False, figsize=(12,6), cmap='coolwarm' )
plt.xlabel('ALCOHOL_CONSUMPTION')
plt.ylabel('Proportion')
plt.title('LUNG CANCER PROPORTION BY ALCOHOL_CONSUMPTION')
plt.show()
```

```python
# Multi-factor Analysis - Age, Gender, Alcohol Consumption vs Lung Cancer
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Age_group', y='PULMONARY_DISEASE', hue='ALCOHOL_CONSUMPTION', style='GENDER', markers=True, dashes=False)
plt.title('Lung Cancer Rate by Age, Gender, and Alcohol Consumption')
plt.xlabel('Age Group')
plt.ylabel('Lung Cancer Rate')
plt.legend(title='Alcohol Consumption', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Multi-factor Analysis - Age, Gender, Alcohol Consumption vs Lung Cancer
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Age_group', y='PULMONARY_DISEASE', hue='ALCOHOL_CONSUMPTION', ci=None, palette='Set2', edgecolor='black')
plt.title('Lung Cancer Rate by Age, Gender, and Alcohol Consumption')
plt.xlabel('Age Group')
plt.ylabel('Lung Cancer Rate')
plt.legend(title='Alcohol Consumption', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix['PULMONARY_DISEASE'].sort_values(ascending=False))
```

```python
# Heatmap Visualization
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## 3.  Health Indicators:

```python
# Health Indicators vs Lung Cancer
health_indicators = ['EXPOSURE_TO_POLLUTION', 'BREATHING_ISSUE', 'CHEST_TIGHTNESS', 'LONG_TERM_ILLNESS']
plt.figure(figsize=(16, 10))
for i, indicator in enumerate(health_indicators, 1):
    plt.subplot(2, 2, i)
    sns.barplot(x=indicator, y='PULMONARY_DISEASE', data=df, ci=None, palette='Set2', edgecolor='black')
    plt.title(f'{indicator} vs Lung Cancer')
    plt.xlabel(indicator.replace('_', ' ').title())
    plt.ylabel('Lung Cancer Rate')
    plt.grid(True, linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

The visualization shows four grouped bar plots comparing different health indicators against lung cancer rates. Each subplot represents a unique factor:

1. Exposure to Pollution vs Lung Cancer: Patients exposed to pollution (1) show a higher lung cancer rate compared to those not exposed (0).

2. Breathing Issue vs Lung Cancer: Patients with breathing issues (1) exhibit a significantly higher lung cancer rate than those without (0).

3. Chest Tightness vs Lung Cancer: Lung cancer rates are slightly higher in patients with chest tightness (1) compared to those without (0).

4. Long Term Illness vs Lung Cancer: The lung cancer rate is similar between patients with and without long-term illness, with a slight increase for those without (0).

```python
health_indicators = ['IMMUNE_WEAKNESS', 'THROAT_DISCOMFORT', 'MENTAL_STRESS', 'FINGER_DISCOLORATION']
plt.figure(figsize=(16, 10))
for i, indicator in enumerate(health_indicators, 1):
    plt.subplot(2, 2, i)
    sns.barplot(x=indicator, y='PULMONARY_DISEASE', data=df, ci=None, palette='Set2', edgecolor='black')
    plt.title(f'{indicator} vs Lung Cancer')
    plt.xlabel(indicator.replace('_', ' ').title())
    plt.ylabel('Lung Cancer Rate')
    plt.grid(True, linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

Older vs. Younger Patients' Symptoms:

Common Symptoms in Lung Cancer Patients:

- Indicators with higher bars for the "Yes" group (Lung Cancer = 1) suggest those symptoms are more common in lung cancer patients.
